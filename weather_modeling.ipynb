Implement weather modeling* using the quadratic solution in stages: hard-coding variables

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def read_coefficients(file_path):
    with open(file_path, "r") as file:
        content = file.readlines()
        coefficients = [line.strip().split(",") for line in content]
        return [[float(coef) for coef in line] for line in coefficients]

def quadratic(time, coefficients):
    a, b, c = coefficients
    temperature = a * (time**2) + b * time + c
    return temperature

def main():
    coefficients_list = read_coefficients("/content/File2.txt")

    for coefficients in coefficients_list:
        time_values = np.linspace(0, 10, 50)
        temp_values = quadratic(time_values, coefficients)

        sns.set_style("darkgrid")
        plt.plot(time_values, temp_values, label=f'Temperature (coefficients: {coefficients})')

    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.title('Quadratic Temperature over Time')
    plt.legend(["set1[34,56,1]","set2[3.4,5.6,7]","set3[0.1,4.0,3]"])
    plt.show()

if __name__ == "__main__":
    main()
```

```python
#Implement weather modeling* using the quadratic solution in stages: hard-coding variables read from a file

import matplotlib.pyplot as plt
import numpy as np

def quadratic_temperature_model(time, a, b, c):
    # Quadratic equation: T(t) = at^2 + bt + c
    temperature = a * (time ** 2) + b * time + c
    return temperature

# List of coefficient sets (a, b, c)
coefficients_list = [
    (0.02, 1.5, 20),
    (0.01, 2, 18),
    (0.03, 1, 25),
]

# Generate time values from 0 to 50 with step 1
time_values = np.arange(0, 51, 1)

# Plotting the results for each set of coefficients
for i, (a, b, c) in enumerate(coefficients_list):
    temperature_values = quadratic_temperature_model(time_values, a, b, c)
    label = f'Set {i + 1}: a={a}, b={b}, c={c}'
    plt.plot(time_values, temperature_values, label=label)

# Plotting details
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Quadratic Temperature Model with Multiple Sets of Coefficients')
plt.legend()
plt.grid(True)
plt.show()
```
